```python
# Step 1: Identify NaN values in the 'title' column
nan_indices = data_file['TITLE'].isna()

# Step 2: Find the most frequent 'title' for each 'cuisine'
most_frequent_titles = data_file.groupby('CUISINES')['TITLE'].agg(lambda x: x.mode().iloc[0] if not x.mode().empty else None)

# Display the most frequent titles for each cuisine
print("\nMost frequent titles by cuisine:")
print(most_frequent_titles)

# Step 3: Fill NaN values in the 'title' column with the most frequent value based on 'cuisine'
data_file.loc[nan_indices, 'TITLE'] = data_file.loc[nan_indices, 'CUISINES'].map(most_frequent_titles)
```

```python
import re

# Correct city names dataset
correct_names = ['Bangalore', 'Thane', 'Hyderabad', 'Andheri', 'Delhi', 'Kerala', 'Chennai', 'Bandra', 'Mumbai', 'Telangana', 'Kochi', 'Noida', 'Gurgaon'
                 , 'Ernakulam', 'Faridabad', 'Ghaziabad', 'Secunderabad','Not Specified']

# Function to match names
def match_names(wrong_name, correct_names):
    for name in correct_names:
        # Constructing a regular expression pattern to match any part of the correct name
        pattern = re.compile(r'\b{}\b'.format(re.escape(name)), flags=re.IGNORECASE)
        if re.search(pattern, wrong_name):
            return name
    # If no match is found, return 'Others'
    return 'Others'

# Call the match_names function to match the names
name_match = [match_names(wrong_name, correct_names) for wrong_name in data_file['LOCATION']]

print(len(name_match))

# Assign the corrected names back to new city the DataFrame
data_file['LOCATION'] = name_match
```

```python
#spliting votes values
votes_values=data_file['VOTES'].str.split(' ',expand=True)
```

```python
data_file['RATING']=data_file['RATING'].replace('-','0').replace('NEW','0').astype(float)
```

```python
#filling nan values with minimum values
data_file['RATING'].fillna(data_file['RATING'].min(), inplace=True)
```

```python
# Function to split and clean the TIME column
def split_time_days(value):
    # Strip leading and trailing whitespaces
    value = value.strip()
    
    # Extract the days part within parentheses
    days_match = re.search(r'\((.*?)\)', value)
    days = days_match.group(1).strip() if days_match else ""
    
    # Remove the days part from the value and strip whitespace
    time_range = re.sub(r'\s*\(.*?\)', '', value).strip()
    
    # Handle cases with "Closed"
    if "Closed" in value:
        time_range = "Closed"
    
    return time_range, days

# Apply the function to split the TIME column
data_file[['OPEN & CLOSE TIME', 'DAYS']] = data_file['TIME'].apply(split_time_days).tolist()
```

```python
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
#find the best random State
max_accuracy=0
best_state=0
random_state=range(1,200)
for state in random_state:
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=state)
    lr=LinearRegression()
    lr.fit(x_train,y_train)
    pred=lr.predict(x_test)
    acc=r2_score(y_test,pred)
    if acc>max_accuracy:
        max_accuracy=acc
        best_state=state
print("Maximum R2_Score is=>",max_accuracy,"On Random State=>",best_state)
```
